```python
import os

import pandas as pd
import networkx as nx
```

```python
preds_100 = pd.read_csv('results//preds100.txt', sep=' ', header=None)
preds_100 = preds_100[0:219]
edges_100 = pd.read_csv('results//edges100.txt', sep=' ', header=None)
```

```python
g = nx.read_edgelist('combined_network.txt')
```

```python
nodes= list(g.nodes)
```

```python
n1=[]
n2=[]
for i in range(edges_100.shape[0]):
    n1.append(nodes[int(edges_100[0][i])])
    n2.append(nodes[int(edges_100[1][i])])
    
```

```python
predicted_edges100 = pd.DataFrame(columns = ["a", "b", "pred"])
predicted_edges100['a'] = n1
predicted_edges100['b'] = n2
predicted_edges100['pred'] = list(preds_100[0])
```

```python
edgelist = pd.DataFrame(g.edges)
edgelist["e"] = edgelist[0]+edgelist[1]
e=list(edgelist["e"])
e
```
